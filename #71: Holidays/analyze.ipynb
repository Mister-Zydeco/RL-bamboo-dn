```python
def date_gen():
    curr_date = dt.datetime.strptime('2010-01-01', '%Y-%m-%d')
    while curr_date.year < 2025:
        yield curr_date
        curr_date += dt.timedelta(days=1)
keys = ['country name', 'alpha2', 'date', 'holiday name']
col_dict = {k: [] for k in keys}
for country in pyc.countries:
    try:
        cholidays = hld.country_holidays(country.alpha_2)
    except NotImplementedError as nie:
        pass
    else:
        for date in date_gen():
            if date in cholidays:
                col_dict['country name'].append(country.name)
                col_dict['alpha2'].append(country.alpha_2)
                col_dict['date'].append(date)           
                col_dict['holiday name'].append(cholidays.get(date))
df = pd.DataFrame.from_dict(col_dict)
df
```
