```python
import pandas as pd
import numpy as np
import pyarrow as pa
import tabula as tb

dfs = tb.read_pdf('./data/80years_2023.pdf', pages=1)
df = dfs[0].iloc[:-3, :-1]
df.columns = [x if ix == 0 else f'{x} {df.iloc[0, ix]}'
              for ix, x in enumerate(df.columns)]

df = (
    df.drop(0, axis='index').dropna(axis='index', how='all')
      .replace('[,$]', '', regex=True)
)
#df['Year'] = pd.to_datetime(df['Year'], '%Y')
coltypes = {x: pd.Int16Dtype() for x in df.columns if x != 'Year'}
coltypes['All Hazard Damages (M)'] = pd.Float64Dtype()
coltypes['Year'] = int
df = df.astype(dtype=coltypes).set_index('Year')
[(col, df[col].first_valid_index()) for col in df.columns]
```

```python
df = df.fillna(0)
df1 = df['Heat Fatalities'].to_frame()

df1['TFH'] = (
        df['Tornado Fatalities'] + df['Flood Fatalities']
        + df['Hurricane Fatalities']
)

df1.loc[1986:, :]
```
