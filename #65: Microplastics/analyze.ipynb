```python
import pandas as pd
import numpy as np
from collections import Counter

df = pd.read_csv('./data/Marine_Microplastics_WGS84_-8073702636451612242.csv', parse_dates=['Date'])
df
```

```python
(df.groupby(lambda x: df['Date'].loc[x].year).count()['OBJECTID']
   .rename('Number of mesurments taken')
   .plot()
 )
```

```python
df.groupby(lambda x: df['Date'].loc[x].year).count()['OBJECTID'].rename(columns={'OBJECTID': 'Nu'})
```

```python
df.sort_values(by='Latitude').iloc[np.r_[:10, -10:0]]
```

```python
(df.query('Unit == "pieces/m3"').groupby(lambda x: df['Date'].loc[x].year)
   .count()['OBJECTID'].pct_change().plot())
```

```python
(df.query('Unit == "pieces/m3"')
   .loc[:, ['Regions', 'Measurement']].groupby('Regions').max()
   .sort_values(by='Measurement', ascending=False)
)
```

```python
(df.query('Unit == "pieces/m3" and Regions.notna()')
   .loc[:, ['Regions', 'Measurement']].sort_values(by='Measurement', ascending=False)
)
```
