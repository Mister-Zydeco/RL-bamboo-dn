```python
import numpy as np
import pandas as pd
import duckdb

df = pd.read_excel('./data/idea_export_voter_turnout_database_region.xlsx', sheet_name='All')
df.memory_usage(deep=True)
```

```python
df
```

```python
conversions = {
    'Year': 'datetime64[ns]', 'Voter Turnout': np.float64, 'Total vote': np.float64, 
    'Registration': np.float64, 'VAP Turnout': np.float64, 'Voting age population': np.float64,    
    'Population': np.float64, 'Invalid votes': np.float64, 'Compulsory voting': 'category',
    'Election Type': 'category'
}
df1 = df.replace('[,%]', '', regex=True).astype(conversions)
df1.info()
```

```python
df1.memory_usage(deep=True)
```

```python
df1.query('`Compulsory voting` == "No"')
```

```python
df1['Election Type'].drop_duplicates()
duckdb.execute('CREATE TABLE df1_duck AS SELECT * FROM df1')
con = duckdb.connect(':default:')
```

```python
%%timeit
(df1.query('`Compulsory voting` == "No"')
     .loc[:, ['Country', 'VAP Turnout']].groupby(['Country']).mean()
     .sort_values(by='VAP Turnout', ascending=False)
 )
```

```python
%%timeit
con.sql('''
    SELECT Country, AVG("VAP Turnout") AS avap FROM df1_duck
    WHERE "Compulsory voting" = 'No'
    GROUP BY Country
    ORDER BY avap DESC'''
)
```

```python
(df1.loc[:, ['Country', 'Compulsory voting', 'Invalid votes']]
    .groupby(['Country', 'Compulsory voting'], observed=True).agg(
        **{'Mean': ('Invalid votes', 'mean'), 'Median': ('Invalid votes', 'median')}
    )
)
```

```python
con.sql('''
SELECT Country, "Compulsory voting", AVG("Invalid votes"), Median("Invalid votes")
FROM df1_duck
WHERE "Compulsory voting" IS NOT NULL
GROUP BY Country, "Compulsory voting"
''')
```
